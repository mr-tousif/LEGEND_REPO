- In the light model, we have the exponential function:

\begin{equation*}
\begin{split}
f(t) &= e^{\frac{-t}{\tau}}, \,\,\,t\geq 0\\
 & = 0, \,\,\,t\leq 0
\end{split}
\end{equation*}

- Our goal is to calculate the exponnential norm fucntion out of the light model we have: To do so we may start with the model and we will convolute it with the gaussian distribution results,

\begin{align*}
\tilde f(t)=\int_{0}^{\infty} \frac{1}{\sqrt{2\pi}\sigma}e^{-\frac{1}{2}\left(\frac{t^\prime-t}{\sigma}\right)^2}e^{\frac{-t^{\prime}}{\tau}} dt^{\prime}\tag{10}
\end{align*}

In the above $t^{\prime}$ is the integration variables and for sake of simplicity we re-define the integration variables as following.

Let,  $t^{\prime}-t=x$ then $dt^{\prime}=dx$. When $t^{\prime}=0, x=-t$ and at $t^{\prime}=\infty, x=\infty$.
eqn.(10) becomes,

\begin{align*}
\tilde f(t)=\frac{1}{\sqrt{2\pi}\sigma}\int_{-t}^{\infty} e^{-\frac{x^2}{2\sigma^2} }e^{\frac{-(x+t)}{\tau}} dx\\
\tilde f(t)=\frac{1}{\sqrt{2\pi}\sigma}e^{\frac{-t}{\tau}}\int_{-t}^{\infty} e^{-\frac{x^2}{2\sigma^2} }e^{\frac{-x}{\tau}} dx\\
\tilde f(t)=\frac{1}{\sqrt{2\pi}\sigma}e^{\frac{-t}{\tau}}\int_{-t}^{\infty} e^{-\frac{1}{2\sigma^2}\left(x^2+\frac{2\sigma^2}{\tau}x\right)}dx\tag{11}
\end{align*}

In the eqn.(11) we have taken $e^{\frac{-t}{\tau}}$ outside of the integration since this is a constant factor does not depend on the integrand which is a function of x only.

To do the above integration we re-write eqn.(11) in some convenient form as follows:

$$\left(x+\frac{\sigma^2}{\tau}\right)^2=x^2+2x\frac{\sigma^2}{\tau}+\frac{\sigma^4}{\tau^2}$$
This indicates, $x^2+2x\frac{\sigma^2}{\tau}=\left(x+\frac{\sigma^2}{\tau}\right)^2-\frac{\sigma^4}{\tau^2} $. Putting back to eqn.(11) this gives,

\begin{align*}
\tilde f(t)=\frac{1}{\sqrt{2\pi}\sigma}e^{\frac{-t}{\tau}}\int_{-t}^{\infty} e^{-\frac{1}{2\sigma^2}\left(\left(x+\frac{\sigma^2}{\tau}\right)^2-\frac{\sigma^4}{\tau^2} \right)}dx\tag{12}\\
\tilde f(t)=\frac{1}{\sqrt{2\pi}\sigma}e^{\frac{-t}{\tau}}e^{\frac{\sigma^2}{2\tau^2}}\int_{-t}^{\infty} e^{-\frac{1}{2\sigma^2}{\left(x+\frac{\sigma^2}{\tau}\right)^2}dx}\tag{13}
\end{align*}


Once again we use re-definition of varibale to deal with the eqn.(13). We set $$x+\frac{\sigma^2}{\tau}=u$$
$$dx=du$$
Derivative of $\frac{\sigma^2}{\tau}$ is zero since $\frac{\sigma^2}{\tau}$ is a constant term. 


Change of limit: when $x=-t, \,\,u=-t+\frac{\sigma^2}{\tau}$ and at $x=\infty, u=\infty$. Putting all these in the eqn.(13) we obtained,
\begin{align*}
\tilde f(t)=\frac{1}{\sqrt{2\pi}\sigma}e^{\frac{-t}{\tau}}e^{\frac{\sigma^2}{2\tau^2}}\int_{-t+\frac{\sigma^2}{\tau}}^{\infty} e^{-\frac{u^2}{2\sigma^2}}du\tag{14}
\end{align*}

One more time we re-define the integration variable in eqn.(14) to get in a simplest form:
$${\frac{u}{\sqrt{2}\sigma}}=v$$
$${\frac{u^2}{2\sigma^2}}=v^2$$
$$du={\sqrt{2}\sigma} dv$$
Change of limit: when when $u=-t+\frac{\sigma^2}{\tau},\,\, v=\frac{1}{\sqrt{2}\sigma}(-t+\frac{\sigma^2}{\tau})$ and at $u=\infty, v=\infty$. Now withh all these the eqn.(14) takes the form,

\begin{align*}
\boxed{\tilde f(t)=\frac{1}{\sqrt{2\pi}\sigma}e^{\frac{-t}{\tau}}e^{\frac{\sigma^2}{2\tau^2}}\sqrt{2}\sigma\int_{\frac{-t+\frac{\sigma^2}{\tau}}{\sqrt{2}\sigma}}^{\infty} e^{-v^2}dv}\tag{14}
\end{align*}

Eqn.(14) is called Exponnetial norm function or Gaussian smeared fuction or Exponentially modified gaussian function. Details below,

https://en.wikipedia.org/wiki/Exponentially_modified_Gaussian_distribution

Complementary error function has the form:

$$erfc(X)=1-erf(X)=\frac{2}{\sqrt{\pi}}\int_{X}^{\infty} e^{-t^2}dt\tag{14*}$$

That means $$\int_{X}^{\infty} e^{-t^2}dt=\frac{\sqrt{\pi}}{2}erfc(X)=\frac{\sqrt{\pi}}{2}\left[1-erf(X)\right]\tag{15}$$

This density function is derived via convolution of the normal and exponential probability density functions.

Then comparing eqn.(14) and eqn.(15), it is straightforward to write the answer:

\begin{align*}
\tilde f(t)=\frac{1}{\sqrt{2\pi}\sigma}e^{\frac{-t}{\tau}}e^{\frac{\sigma^2}{2\tau^2}}\sqrt{2}\sigma\frac{\sqrt{\pi}}{2}erfc \left(\frac{-t+\frac{\sigma^2}{\tau}}{\sqrt{2}\sigma}\right)\tag{16}
\end{align*}
Finally we have,
\begin{align*}
\tilde f(t)=\frac{1}{2}e^{\frac{-t}{\tau}}e^{\frac{\sigma^2}{2\tau^2}}\,\,erfc \left(\frac{-t+\frac{\sigma^2}{\tau}}{\sqrt{2}\sigma}\right)\tag{17}\\
\boxed{\tilde f(t)=\frac{1}{2}e^{\frac{-t}{\tau}}e^{\frac{\sigma^2}{2\tau^2}}\left[1-erf \left(\frac{-t+\frac{\sigma^2}{\tau}}{\sqrt{2}\sigma}\right)\right]}\tag{18}
\end{align*}

Eqn.(18) is the desired exponential norm function or smeared gaussian function that we are going to implement for the fitting.

However if we take the model in a form: 
\begin{equation*}
\begin{split}
f(t) &= \frac{1}{\tau}e^{\frac{-t}{\tau}}, \,\,\,t\geq 0\\
 & = 0, \,\,\,t\leq 0
\end{split}
\end{equation*}

Then the the desired exponential norm function will have $\frac{1}{\tau}$ extra factor. Means the final expressions would be,

\begin{align*}
\boxed{\tilde f(t)=\frac{1}{2\tau}e^{\frac{-t}{\tau}}e^{\frac{\sigma^2}{2\tau^2}}\left[1-erf \left(\frac{-t+\frac{\sigma^2}{\tau}}{\sqrt{2}\sigma}\right)\right]}\tag{19}
\end{align*}


- Comments: Eqn.(18) and Eqn.(19) is our analytical calculations consistent with the exponential gaussian that is inbuilt in the cern-root library. However we can take a look at the python version of the expontial gausssian(which is nothing but exponnorm function in the scipy library).

In the above expression we just multiplied and devide by a sigma in the parenthesis. Now we set $x\sigma=y$ then eqn. (24) takes the form,

\end{align*}

The recombination term is given by:

$f_{rec}(t)=\frac{A_{rec}}{\left(1+\frac{t}{\tau_{rec}}\right)^2}$

The convolution of recombination term is given by,

\begin{align*}
\tilde f_{recomb}(t)=\int_{0}^{\infty} \frac{1}{\sqrt{2\pi}\sigma}e^{-\frac{1}{2}\left(\frac{t^\prime-t}{\sigma}\right)^2} \frac{A_{rec}}{\left(1+\frac{t}{\tau_{rec}}\right)^2} dt^{\prime} \tag{11}
\end{align*}

The integration in eqn.(11) can be solved by Mathematica taking the Taylor expansion of $f_{rec}$ upto certain order.  Here I have considered upto $O(3)$.

```python
import numpy as np
from scipy import special
import matplotlib.pyplot as plt
```

```python
bw=2
norm=9.8*1e4
sfrac=0.22
rfrac=0.05 # taking 5% probability of having recombination in the liquid Argon
alpha1=sfrac * bw * norm;
alpha3=(1-sfrac)*bw*norm
```

```python
a=100 #t in the Gaussian function
sigma=10
b=1/(2*sigma**2)
A_rec=1
tau_rec=37.1 #ns


A_rec=(1-sfrac-rfrac)*bw*norm

def zeroth_term(a,b,x):
    factor_1=np.sqrt(np.pi)/(2*np.sqrt(b))
    f_1=factor_1*(special.erf(a*np.sqrt(b))+1) #constant
    return A_rec*f_1
    
def first_term(a,b,x):
    factor_2=1/(2*b)
    f_2=(np.sqrt(np.pi)*a*np.sqrt(b)*(special.erf(np.sqrt(b)*(a-x))+1)+np.exp(-b*(a-x)**2))/factor_2
    return A_rec*f_2

def second_term(a,b,x):
    factor_3=1/(4*b**1.5)
    f_3=(np.sqrt(np.pi)*(2*a**2*b+1)*(special.erf(a*np.sqrt(b))+1)/factor_3)+((a/(2*b))*np.exp(-a**2*b))
    return A_rec*f_3

def third_term(a,b,x):
    factor_4=1/(4*b**2)
    f_4=np.sqrt(np.pi)*a*np.sqrt(b)*(2*a**2*b+3)*(special.erf(a*np.sqrt(b))+1)+2*np.exp(-a**2*b)*(a**2*b+1)/factor_4
    return A_rec*f_4

    
def f_recomb(a,b,x):
    factor_1=np.sqrt(np.pi)/(2*np.sqrt(b)) 
    f_1=factor_1*(special.erf(a*np.sqrt(b))+1)
    factor_2=1/(2*b) 
    f_2=(np.sqrt(np.pi)*a*np.sqrt(b)*(special.erf(np.sqrt(b)*(a-x))+1)+np.exp(-b*(a-x)**2))/factor_2
    factor_3=1/(4*b**1.5)
    f_3=(np.sqrt(np.pi)*(2*a**2*b+1)*(special.erf(a*np.sqrt(b))+1)/factor_3)+((a/(2*b))*np.exp(-a**2*b))
    factor_4=1/(4*b**2)
    f_4=np.sqrt(np.pi)*a*np.sqrt(b)*(2*a**2*b+3)*(special.erf(a*np.sqrt(b))+1)+2*np.exp(-a**2*b)*(a**2*b+1)/factor_4
    A_rec=(1-sfrac-rfrac)*bw*norm
    return (A_rec/np.sqrt(2*np.pi*sigma**2))*(f_1-(2/tau_rec)*f_2+(3/tau_rec**2)*f_3-(4/tau_rec**3)*f_4) #considering term upto order(3)

```

```python
x=np.linspace(0,100,100) # time in ns
```

```python
zeroth_term=zeroth_term(a,b,x) #constant term
print(zeroth_term)
```

```python

first_term=first_term(a,b,x) #has time dependence
print(first_term)
```

```python
second_term=second_term(a,b,x) #constant term

print(second_term)
```

```python
third_term=third_term(a,b,x)
print(third_term)
```

```python
f_recomb=f_recomb(a,b,x)
```

```python
plt.plot(x,first_term, color="green")
plt.xlabel(r"$t\,[ns]$")
plt.ylabel("counts")
plt.show()
```

```python
plt.plot(x,f_recomb, color="purple")
plt.xlabel(r"$t\,[ns]$")
plt.ylabel("counts")
plt.show()
```

```python
#plotting the singlet and Triplet term

from scipy.stats import exponnorm

y=x*sigma

def Singlet_c(a,b,x):
    tau_s=6 #ns
    lambda_s=1/tau_s
    K_s=1/(sigma*lambda_s)
    return alpha1*K_s*exponnorm.pdf(y,K_s)
```

```python
Singlet_c=Singlet_c(a,b,x)
```

```python
plt.plot(x,Singlet_c, color="black", lw=2, label="Singlet")
plt.ylabel("count")
plt.xlabel("t[ns]")
plt.legend()
```

```python
def Triplet_c(a,b,x):
    tau_T=1500 #ns
    lambda_T=1/tau_T
    K_T=1/(sigma*lambda_T)
    return alpha3*K_T*exponnorm.pdf(y,K_T)
```

```python
Triplet_c=Triplet_c(a,b,x)

plt.plot(x,Triplet_c, color="blue", lw=2, label="Triplet")
plt.xlabel("t[ns]")
plt.ylabel("counts")
plt.legend()
```

```python

plt.plot(x, Singlet_c, color="magenta", lw=3)
plt.plot(x, Triplet_c, color="cyan", lw=3)
plt.plot(x, f_recomb, color="blue", lw=3)
plt.xlabel("t[ns]")
plt.ylabel("counts")
plt.xlabel("t[ns]")
plt.ylabel("counts")
plt.xlabel("t[ns]")
plt.ylabel("counts")
```

```python
# full contribution to the model

plt.plot(x, Singlet_c+Triplet_c+f_recomb, color="magenta", lw=3, label=r"$f_s+f_t+f_{rec}$")
plt.xlabel("t[ns]")
plt.ylabel("counts")
plt.legend()

```
